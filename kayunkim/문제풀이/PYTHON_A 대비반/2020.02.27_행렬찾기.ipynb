```python
#1258. [S/W 문제해결 응용] 7일차 - 행렬찾기
#단순탐색
T = int(input())
for tc in range(1,T+1):
    N = int(input())
    mtx = [list(map(int,input().split())) for _ in range (N)]
    for i in range(N):
        cnt = 0
        for j in range(N):
            if mtx[i][j] != 0 :
                cnt+=1
                if j != 0:
                    mtx[i][j-1] = 0
                mtx[i][j] = cnt
            else:
                cnt = 0

    dic = {}
    for j in range(N):
        cnt = 0
        for i in range(N):
            if mtx[i][j] != 0:
                cnt += 1
                if i == N-1:
                    dic[cnt] = mtx[i][j]
                    break
            else:
                if cnt != 0:
                    dic[cnt] = mtx[i-1][j]
                cnt = 0
    bin = []
    for i in dic:
        bin.append(i)
        bin.append(dic[i])
    for _ in range(len(bin)//2-1):
        maxV = 0
        for i in range(0,len(bin),2):
            if maxV < bin[i]*bin[i+1]:
                maxV = bin[i]*bin[i+1]
            elif maxV == bin[i]*bin[i+1]:
                if bin[i-1] > bin[i]:
                    bin[i],bin[i-2] = bin[i-2], bin[i]
                    bin[i+1], bin[i-1] = bin[i-1], bin[i+1]
            else:
                bin[i],bin[i-2] = bin[i-2], bin[i]
                bin[i+1], bin[i-1] = bin[i-1], bin[i+1]

    print('#{} {} {}'.format(tc,len(dic),' '.join(map(str,bin))))
```
