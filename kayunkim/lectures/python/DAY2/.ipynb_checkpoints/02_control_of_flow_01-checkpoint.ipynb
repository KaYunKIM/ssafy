# 구구단 출력하기
> 2단부터 9단까지 반복문을 사용하여 구구단을 출력하세요.
---
```
예시 출력)
------- [2 단] -------
2 X 1 = 2
2 X 2 = 4
2 X 3 = 6
2 X 4 = 8
2 X 5 = 10
2 X 6 = 12
2 X 7 = 14
2 X 8 = 16
2 X 9 = 18

------- [3 단] -------
...
```

````python
# 아래에 코드를 작성하세요.
````

````python

for x in range(2, 10):
    print(f'---- {x}단 ----')
    for y in range(1,10):
        print(f'{x} X {y} = {x*y}')
    print()
````

# 달력 출력하기

> 1월 1일 월요일부터 12월 31일까지 달력을 출력하세요.

---
```

         3 월

 1  2  3  4  5  6  7

```

````python
calendar = { 
        1: 31, 2: 28, 3: 31, 4: 30, 5: 31,6: 30, 
        7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31 
} 
weeks = ['Mo', 'Tu', 'We', 'Th', 'Fr', 'Sa', 'Su'] 

count = 0

for m, d in calendar.items():
    print(f'{m:10}월')
    count = 0
    for w in weeks:
        print(f'{w:3}', end = '')
    print()
    for i in range(1,d+1):
        count += 1
        print(f'{i:2}', end=' ')
        if count %7 == 0:
            print()
    print(); print()


# 아래에 코드를 작성하세요.
````

````python
for month, count_day in calendar.items():
    count =0
    print(f'{month:10}', '월') #10칸안에 달을 출력해라
    for day in weeks:
        print(f'{day}', end =' ')
    print()
        
    for i in range(1, count_day+1):
        print(f'{i:2}', end = ' ')
        count += 1
        if count == 7:
            print()
            count= 0
    print()
````

# 정중앙
> 사용자가 입력한 문자열중 가운데 글자를 출력하세요. 
>
> 단, 문자열이 짝수라면 가운데 두글자를 출력하세요.

---

```
예시 입력)
abc

예시 출력)
b
```

````python
# 아래에 코드를 작성하세요.
````

````python
a = input()
if len(a) %2 == 1:
    print(a[len(a)//2])
else:
    print(a[(len(a)//2)-1] + a[len(a)//2])

# a= input()
# print(a[(len(a)//2)-1:len(a)//2+1])
````

````python
text = input()
num = len(text) //2

if len(text) % 2:
    middle = text[num]
else:
    middle = text[num-1:num+1]
print(middle)
````

# 소수 찾기

> 조건, 반복문을 응용하여 numbers 리스트의 요소들이 소수인지 아닌지 판단하는 코드를 작성하세요.

---

예시)

```python
numbers = [26, 39, 51, 53, 57, 79, 85]

26 는 소수가 아닙니다. 2 는 26 의 인수입니다.
39 는 소수가 아닙니다. 3 는 39 의 인수입니다.
51 는 소수가 아닙니다. 3 는 51 의 인수입니다.
53 는 소수입니다.
57 는 소수가 아닙니다. 3 는 57 의 인수입니다.
79 는 소수입니다.
85 는 소수가 아닙니다. 5 는 85 의 인수입니다.
```

````python
numbers = [26, 39, 51, 53, 57, 79, 85]

# 아래에 코드를 작성하세요.
````

````python
for i in numbers:
    is_prime = True
    # 2~n-1 나눠 떨어지는가 체크 반복
    for x in range(2, i):
        if i% x != 0:
            pass
        else:
            # 2~n-1 나눠떨어진 것이 있다. -> 소수가 아니게 된다.
            is_prime = False
            print(f'{i}는 소수가 아닙니다. {x}는 {i}의 인수입니다.')
            break 
    if is_prime:
        print(f'{i}는 소수입니다.')     
            
            
            
````
