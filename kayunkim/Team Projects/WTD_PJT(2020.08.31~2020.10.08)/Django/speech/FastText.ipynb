```python
from gensim import models #fasttext 라이브러리.
```

```python
ko_model = models.fasttext.load_facebook_model('cc.ko.300.bin.gz') #fasttext 한국어 pre trained 모델 로드하기.
```

```python
for w, sim in ko_model.wv.similar_by_word('자바', 10): #pre trained 된 모델에서 특정 단어와 가장 가까운 단어들 찾기.
    print(f'{w}: {sim}')
```

```python
from sklearn.decomposition import PCA #차원 축소 라이브러리. 벡터를 2차원으로 줄여주는 역할을 하여 시각화 쉽게 해준다.
import matplotlib.pylab as plt #시각와를 위한 라이브러리.

words = [
    '사과',
    '바나나',
    '오렌지',
    '키위',
    '스파이더맨', 
    '아이언맨', 
    '헐크',
    '타노스',
    '캡틴아메리카',
    '어벤져스'
]
# matplotplib 폰트 설정을 안해서 그래프에서 한국어 라벨이 깨져서 아래 단어로 임시처리했습니다. ㅠ
word_labels = [
    'apple',
    'banana',
    'orange',
    'kiwi',
    'spider man', 
    'iron man', 
    'hulk',
    'thanos',
    'captin america',
    'avengers'
]
```

```python
pca = PCA(n_components=2) #차원축소
xys = pca.fit_transform([ko_model.wv.word_vec(w) for w in words]) #모델내 특정 단어들 벡터 뽑아 차원축소
xs = xys[:,0] #차원축소 x축.
ys = xys[:,1] #차원축소 y축.
plt.figure(figsize=(14, 10)) #시각화 크기.
plt.scatter(xs, ys, marker='o') # 시각화 축 정하기.
for i, v in enumerate(word_labels): 
    plt.annotate(v, xy=(xs[i], ys[i])) #라벨링별 점찍기
```
